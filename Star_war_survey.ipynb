Converting each column to a Boolean having only the values True, False

```python
yes_no = {
    "Yes": True,
    "No": False
}
```

```python
for c in ['Have you seen any of the 6 films in the Star Wars franchise?', 'Do you consider yourself to be a fan of the Star Wars film franchise?']:
    star_wars[c] = star_wars[c].map(yes_no)
```

```python
star_wars.head()
```

The next six columns represent a single checkbox question. The respondent checked off a series of boxes in response to the question, **Which of the following Star Wars films have you seen? Please select all that apply**.  

For each of these columns, if the value in a cell is the name of the movie, that means the respondent saw the movie. If the value is NaN, the respondent either didn't answer or didn't see the movie. We'll assume that they didn't see the movie.  

Let's converting each column to a Boolean like we did it on previous step.

```python
episodes = {'Star Wars: Episode I  The Phantom Menace':True,
            np.nan: False,
            'Star Wars: Episode II  Attack of the Clones': True,
            'Star Wars: Episode III  Revenge of the Sith':True,
            'Star Wars: Episode IV  A New Hope' : True,
            'Star Wars: Episode V The Empire Strikes Back': True,
            'Star Wars: Episode VI Return of the Jedi': True
}
```

```python
for c in star_wars.columns[3:9]:
    star_wars[c] = star_wars[c].map(episodes)
    
star_wars.head()
```

From the name of the columns, like **Unnamed: 4**, it does not clear what information its contain. Let's rename appropriate columns.

```python
star_wars = star_wars.rename(columns={
    "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
    'Unnamed: 4':"seen_2",
    'Unnamed: 5':"seen_3",
    'Unnamed: 6':"seen_4",
    'Unnamed: 7':"seen_5",
    'Unnamed: 8':"seen_6"
})
```

```python
star_wars.head()
```

In next few columns respondents are asked to rate films in the franchise. Let's change type of columns on float an rename appropriate columns.

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
```

```python
star_wars.columns[9:15]
```

```python
star_wars = star_wars.rename(columns={
    'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.': "ranking_1",
    'Unnamed: 10':"ranking_2",
    'Unnamed: 11':"ranking_3",
    'Unnamed: 12':"ranking_4",
    'Unnamed: 13':"ranking_5",
    'Unnamed: 14':"ranking_6"
})
```

It is intresting to understand which film from the franchise is liked more people. This can be understood by calculating the average values of **'ranking'** columns.  
Write a function for this task.

```python
def mean_ranks(data):
    mean_rank = {}
    for c in data.columns[9:15]:
        a = data[c].mean()
        mean_rank[c] = a
    return mean_rank
    
mean_ranks(star_wars)    
```

Let's see how many respondents watched each movie

```python
count_seen = {}
for c in star_wars.columns[3:9]:
    b = star_wars[c].sum()
    count_seen[c] = b
    
count_seen
```

Let' visualise ranking and number of respondents who watch film

```python
num_cols = ['seen_1', 'seen_2', 'seen_3', 'seen_4', 'seen_5', 'seen_6']

bar_heights = star_wars[star_wars.columns[3:9]].sum()
bar_positions = arange(6) + 0.75
tick_positions = range(1,7)

fig, ax = plt.subplots()
ax.bar(bar_positions, bar_heights, 0.5)
ax.set_xticks(tick_positions)
ax.set_xticklabels(num_cols, rotation=90)
plt.xlabel("Episode")
plt.ylabel("Number of Views")
plt.title("Number of Views of Star Wars Episodes")

num_cols_rank = ['ranking_1', 'ranking_2', 'ranking_3', 'ranking_4', 'ranking_5', 'ranking_6']
bar_heights_rank = star_wars[star_wars.columns[9:15]].mean()
bar_positions_rank = arange(6) + 0.75
tick_positions_rank = range(1,7)

fig, ax = plt.subplots()
ax.bar(bar_positions_rank, bar_heights_rank, 0.5)
ax.set_xticks(tick_positions_rank)
ax.set_xticklabels(num_cols_rank, rotation=90)
plt.xlabel("Episode")
plt.ylabel("Rank")
plt.title("Rank of Star Wars Episodes")

plt.show()
```

From the dispersion of ranking, it is clear, that last 3 movies are liked by less number of respondents.  
There is two columns in data, which can help us to understand this.
 - Gender - The respondent's gender  
 - Age - The respondent's age

```python
print(star_wars['Age'].value_counts())
print(star_wars['Gender'].value_counts())
```

There is 4 age groups and the approximate same number of respondents belong to different age groups. The same we see in Gender groups.  
Calculate average rank for every age and gender groups

```python
A = star_wars[star_wars['Age'] == '45-60']
B = star_wars[star_wars['Age'] == '> 60']
C = star_wars[star_wars['Age'] == '30-44']
D = star_wars[star_wars['Age'] == '18-29']
```

```python
age_mean_rank = []
for a in [A,B,C,D]:
    r = mean_ranks(a) 
    age_mean_rank.append(r)

age_mean_rank
```

```python
F = star_wars[star_wars['Gender'] == 'Female']
M = star_wars[star_wars['Gender'] == 'Male']

gender_mean_rank = []
for a in [F,M]:
    r = mean_ranks(a) 
    gender_mean_rank.append(r)

gender_mean_rank
```

Besides, it is interesting to see how the estimates of the episodes were distributed among respondents who consider themselves to fans of the franchise and respodents who not consider.

```python
L = star_wars[star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] == True]
U = star_wars[star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] == False]

fun_mean_rank = []
for a in [L,U]:
    r = mean_ranks(a) 
    fun_mean_rank.append(r)

fun_mean_rank
```

## Conclusion: 
From the analysis it's clear that all respondents like episodes 1-3 more regardless of gender and age. Among the fans of the franchise evaluation between the old (4-6) and the new (1-3) episodes are divided quite clearly: 4.3 for new episodes, against 2.7 for the old. Most likely this is due to the fact that the movies 1-3 are newer and they used more special effects, so they are more spectacular. Respondents who do not consider themselves to fans are more loyal in the evaluation of episodes.
